```python
from bargain_watch_data.image_search import SearchController, SearchAnnotator
from bargain_watch_data.item_page import ItemPageController, ItemPageAnnotator
from selenium import webdriver


driver = webdriver.Firefox()


item_page_controller = ItemPageController()
item_page_controller._driver=driver
item_page_annotator = ItemPageAnnotator(item_page_controller)

search_controller = SearchController()
search_controller._driver=driver
search_annotator = SearchAnnotator(search_controller)
```

# 1. Extract item's images
## 1.1. Set item's url

```python
item_url = "https://www.zalando.pl/esprit-koszula-dark-turquoise-es122d0wa-l11.html"
```

## 1.2. Extract images

```python
item_page_controller.start()
item_page_controller.request(item_url)
item_page_annotator.reset()
```

## 1.3. Select images

```python
item_page_annotator.draw()
```

## 1.4. Save annotated images

```python
item_page_controller.save()
```

# 2. Annotate

## 2.1. Gather data from web

```python
search_controller.start()
```

```python
for image in item_page_controller.item_page.final_images:
    search_controller.request(image.url)
```

## 3. Annotate data

```python
search_annotator.draw()
```

## 4. Save results

```python
search_controller.save()
```
